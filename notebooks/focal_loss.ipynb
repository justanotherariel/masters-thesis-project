```python

import numpy as np
import matplotlib.pyplot as plt

def plot_focal_loss():
    p = np.linspace(0.01, 1, 100)
    
    plt.figure(figsize=(10, 6))
    plt.plot(p, -np.log(p), label='CE Loss')
    plt.plot(p, -(1-p)**1 * np.log(p), label='FL (γ=1)')
    plt.plot(p, -(1-p)**2 * np.log(p), label='FL (γ=2)')
    plt.plot(p, -(1-p)**5 * np.log(p), label='FL (γ=5)')
    # plt.plot(p, -(1-p)**2 * np.log(p*0.5), label='FL Times (γ=5)')
    # plt.plot(p, -(1-p)**10, label='FL - (γ=5)')
    
    plt.xlabel('Predicted Probability for Correct Class')
    plt.ylabel('Loss Value')
    plt.legend()
    plt.grid(True)
    plt.title('Comparison of Cross Entropy vs Focal Loss')
    
    plt.show()
    
plot_focal_loss()
```
